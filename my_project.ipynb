```python
import pandas as pd
import numpy as np

df = pd.read_csv("monthly_revenue.csv")
print(df.head(10))
```

```python
uniqueCommodity = df['Commodity'].unique()
print(uniqueCommodity)

uniqueProduct = df['Product'].unique()
print(uniqueProduct)
```

```python
#df.loc[(df["Commodity"] == "Oil & gas (pre-production)") & (df["Product"].isna()), "Product"] = "Oil & gas (pre-production)"
df["Product"] = df["Product"].fillna(df["Commodity"])
df = df[df["Commodity"] != "Not tied to a commodity"]
df = df[df["Land Class"] == "Federal"]
df = df[df["Land Category"] == "Onshore"]
df = df.dropna(subset=["State"])
df = df.drop(columns=["Land Class", "Land Category", "County", "FIPS Code", "Revenue Type", "Mineral Lease Type", "Offshore Region"])

df["Date"] = pd.to_datetime(df["Date"])
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")


uniqueCommodity = df['Commodity'].unique()
print(uniqueCommodity)

uniqueProduct = df['Product'].unique()
print(uniqueProduct)

uniqueState = df["State"].unique()
print(uniqueState)

print(f'Null data in the Data frame: \n{df.isnull().sum()}')
print(df.dtypes)
```

```python
#create new csv file that is cleaned
#print(df.head(10))
df.to_csv('cleaned_monthly_revenue.csv', index=False)
```

```python
import sqlite3
import os
#imports and create database
try:
    conn = sqlite3.connect('monthly_revenue.db')
    print("Database monthly_revenue.db created successfully")

except Exception as e:
    print(f"Error creating database: {e}")

```

```python
#create table
try:
    cursor = conn.cursor()
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Revenue(
        Date TEXT,
        State TEXT,
        Commodity TEXT,
        Product TEXT,    
        Revenue FLOAT
    )
    ''')

    conn.commit()
    cursor = conn.cursor()
    #clear db
    cursor.execute('DELETE FROM Revenue')

    for _, row in df.iterrows():
        cursor.execute('''
        INSERT INTO Revenue (Date, State, Commodity, Product, Revenue)
        VALUES (?, ?, ?, ?, ?)
        ''', (row['Date'], row['State'], row['Commodity'], row['Product'], row['Revenue']))

    conn.commit()

    #check count
    cursor.execute("SELECT COUNT(*) FROM Revenue")
    count = cursor.fetchone()[0]
    print(f"Total rows: {count}")
    print("\nFirst few records in the database:")
    cursor.execute("SELECT * FROM Revenue LIMIT 10")
    for row in cursor.fetchall():
        print(row)
    
except Exception as e:
    print(f"Error loading: {e}")
    conn.rollback()
```

```python
query = """
    SELECT
        Date,
        Product,
        SUM(Revenue) AS TotalRevenue
    FROM
        Revenue
    GROUP BY
        Date,
        Product
    ORDER BY
        Date
"""

cursor.execute(query)

result = cursor.fetchall()

netdf = pd.DataFrame(result, columns=["Date", "Product", "TotalRevenue"])

print(netdf.head(20))
```

```python
from scipy import stats

length = len(netdf)

revenueZ = stats.zscore(netdf['TotalRevenue'])
netdf['RevenueZ'] = revenueZ
print(f"outliers in Revenue column\n {netdf['RevenueZ'][np.abs(revenueZ)>3]}")

noOutliers = (np.abs(revenueZ)<=3)
dfclean = netdf[noOutliers].reset_index()
del dfclean['RevenueZ']

percentageOutlier = (((length - len(dfclean))/length)*100)
print(f"Percentage of outliers: {percentageOutlier:.2f}")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
onehot_encoder.fit(dfclean[['Date', 'Product']])
onehot_encoded = onehot_encoder.transform(dfclean[['Date', 'Product']])
#onehot_encoded = onehot_encoder.fit_transform(dfclean[['Date', 'Product']])

#df_encoded = pd.get_dummies(dfclean, columns=['Date', 'Product'])
onehot_df = pd.DataFrame(onehot_encoded, columns=onehot_encoder.get_feature_names_out(['Date', 'Product']))
df_final = pd.concat([dfclean['TotalRevenue'], onehot_df], axis=1)
df_final.reset_index(drop=True, inplace=True)

print(df_final)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df_final.drop('TotalRevenue', axis=1)
X_scaled = scaler.fit_transform(X)
y= df_final['TotalRevenue']

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

model = LinearRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
